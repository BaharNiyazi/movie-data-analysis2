```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("movies.csv")
df.head(5)
```

```python
yearly_rating=df.groupby("release_date")["vote_average"].mean()

plt.figure(figsize=(10,5))
yearly_rating.iloc[::5].plot()
plt.title("Average Mivie Rating Over Time")
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.show()
```
